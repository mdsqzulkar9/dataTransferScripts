```python
# choose the submaster file: 
home_dir = os.environ['HOME']
sep = os.sep
git_data_dir = 'Dropbox' + sep + 'gits' + sep + 'data'
th_folder = 'ThOpt'
sub_master = 'sub_master'

sub_master_file_path = home_dir + sep + git_data_dir + sep + th_folder + sep + sub_master
os.listdir(sub_master_file_path)
```

```python
# Choose the master file : choose the most recent one !
master_dir = 'master'
master_file_path = home_dir + sep + git_data_dir + sep + th_folder + sep + master_dir
os.listdir(master_file_path)
```

```python
master_file_name = 'master_06_29_2016_15_00'
sub_master_file_name = 'master_kemal_bkground_traffic_09_13_2016_10_22'
```

```python
master_data = pds.read_csv(master_file_name)
master_data
```

```python
master_data['artificial_bk_traffic'] = 0
master_data
```

```python
submaster_data = pds.read_csv(sub_master_file_name)
submaster_data['artificial_bk_traffic'] = 1
submaster_data
```

```python
# take care some impurites : 
# change Gordon/Oasis, Gordon to Gordon-Oasis
master_data = master_data.replace(['Gordon'],['Gordon-Oasis'])
master_data = master_data.replace(['Gordon/Oasis'],['Gordon-Oasis'])
submaster_data = submaster_data.replace(['Gordon'],['Gordon-Oasis'])
submaster_data = submaster_data.replace(['Gordon/Oasis'],['Gordon-Oasis'])
```

```python
# output file name :
current_date_time = datetime.datetime.now()
current_time_str = current_date_time.strftime('%m_%d_%Y_%H_%M')
output_file = 'master_' + current_time_str

# append data in an existing csv file with header false :
with open(output_file, 'a') as f:
    master_data.to_csv(f, header=True,index=False)
    submaster_data.to_csv(f, header=False,index=False)
```
